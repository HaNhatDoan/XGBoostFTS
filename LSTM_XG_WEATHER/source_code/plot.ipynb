```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Tạo dataframe từ dữ liệu bảng
data = {
    'Model': ['LSXG (ours)', 'Rlinear', 'PatchTST', 'Crossformer', 'TiDE', 'Dlinear'],
    'ECL_MSE': [0.298, 0.219, 0.205, 0.244, 0.251, 0.212],
    'ECL_MAE': [0.376, 0.298, 0.290, 0.334, 0.344, 0.300],
    'ETT_MSE': [0.292, 0.380, 0.381, 0.685, 0.482, 0.442],
    'ETT_MAE': [0.434, 0.392, 0.397, 0.578, 0.470, 0.444],
}

df = pd.DataFrame(data)

# Thiết lập vị trí vẽ
x = np.arange(len(df['Model']))  # vị trí các mô hình
width = 0.18  # độ rộng mỗi cột

# Vẽ biểu đồ
fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(x - 1.5*width, df['ECL_MSE'], width, label='ECL MSE')
rects2 = ax.bar(x - 0.5*width, df['ECL_MAE'], width, label='ECL MAE')
rects3 = ax.bar(x + 0.5*width, df['ETT_MSE'], width, label='ETT MSE')
rects4 = ax.bar(x + 1.5*width, df['ETT_MAE'], width, label='ETT MAE')

# Thêm nhãn
ax.set_ylabel('Metric Value')
ax.set_title('Comparison of Forecasting Models on ECL and ETT')
ax.set_xticks(x)
ax.set_xticklabels(df['Model'], rotation=15)
ax.legend(loc='upper right')
ax.grid(True, linestyle='--', alpha=0.4)

# Hiển thị giá trị trên đỉnh cột
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.3f}', xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', fontsize=8)

for r in [rects1, rects2, rects3, rects4]:
    autolabel(r)

plt.tight_layout()
plt.show()
```
